```python

missing_values = df.isnull().sum()
missing_values[missing_values > 0]
```

## üîç Missing Values Analysis  
‚úÖ **No missing values detected!**  
- This is a **clean dataset**, meaning we don't need imputation strategies.  
- Proceed to **data cleaning** by addressing irrelevant or redundant columns.  


```python

categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    print(f"üü¢ {col}: {df[col].nunique()} unique values")
    print(df[col].unique()[:10])  
    print("-" * 50)
```

## üè∑Ô∏è Unique Values in Categorical Columns  

- **Gender:** **2 values** ‚Üí No issues.  
- **CampaignChannel:** **5 values** ‚Üí Well-defined marketing channels.  
- **CampaignType:** **4 values** ‚Üí Includes all necessary stages of a marketing funnel.  
- **AdvertisingPlatform & AdvertisingTool:** **1 unique value each** ‚Üí **Redundant columns, should be dropped.**  


```python

columns_to_drop = ['CustomerID', 'AdvertisingPlatform', 'AdvertisingTool']
df_cleaned = df.drop(columns=columns_to_drop, axis=1)

print("üìå Columns after dropping:")
print(df_cleaned.columns)
```

## üöÆ Dropping Unnecessary Columns  

The following columns will be dropped as they do not provide useful variance:  
1Ô∏è‚É£ **CustomerID** ‚Üí A unique identifier, not useful for modeling.  
2Ô∏è‚É£ **AdvertisingPlatform** ‚Üí Only one unique value, adds no analytical value.  
3Ô∏è‚É£ **AdvertisingTool** ‚Üí Only one unique value, redundant.  


```python

duplicates = df_cleaned.duplicated().sum()
print(f"üö® Duplicate Rows: {duplicates}")
```

```python
from sklearn.preprocessing import LabelEncoder


categorical_columns = ["Gender", "CampaignChannel", "CampaignType"]


label_encoders = {}

for col in categorical_columns:
    le = LabelEncoder()
    df_cleaned[col] = le.fit_transform(df_cleaned[col])
    label_encoders[col] = le
    
df_cleaned.head()
```

## üî† Label Encoding Categorical Features  

### üîπ Why Label Encoding?  
- Machine learning models work better with numerical data.  
- We encoded **Gender, CampaignChannel, and CampaignType** into numerical values.  
- **One-Hot Encoding** was avoided since these variables have a limited number of unique values.  

### üîç **Encoding Details:**  
- **Gender** ‚Üí `{Female: 0, Male: 1}`  
- **CampaignChannel** ‚Üí `{Social Media: 0, Email: 1, PPC: 2, Referral: 3, SEO: 4}`  
- **CampaignType** ‚Üí `{Awareness: 0, Retention: 1, Conversion: 2, Consideration: 3}`  


```python
import seaborn as sns
import matplotlib.pyplot as plt


corr_matrix = df_cleaned.corr()


plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("üîç Correlation Heatmap of Features")
plt.show()
```

## üî• Correlation Analysis: Heatmap of Features  

### üîç **Key Observations**  
- The correlation values are mostly **low** across features, indicating weak linear relationships.  
- **AdSpend** and **ClickThroughRate** show a **moderate correlation** with **Conversion** (~0.12).  
- **TimeOnSite** and **Email Opens/Clicks** also show a weak **positive correlation** with **Conversion** (~0.13).  
- **Income, Age, and Gender** have almost no impact on **Conversion**.  
- No features exhibit a **strong correlation (>|0.5|)** with each other, meaning **low multicollinearity**.

```python
import seaborn as sns
import matplotlib.pyplot as plt


engagement_metrics = ["WebsiteVisits", "PagesPerVisit", "TimeOnSite", "SocialShares", "EmailOpens", "EmailClicks"]


plt.figure(figsize=(12, 6))
for i, metric in enumerate(engagement_metrics, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x=df_cleaned["Conversion"], y=df_cleaned[metric])
    plt.title(f"{metric} Distribution by Conversion")

plt.tight_layout()
plt.show()
```

### üìä Feature Relationships: Engagement Metrics vs. Conversion

We analyzed the relationship between key engagement metrics and conversion rates using box plots.  

#### üîπ **Insights:**
- **Website Visits**: Users with a higher number of visits tend to have slightly higher conversion rates.
- **Pages Per Visit**: Converted users tend to visit slightly more pages per session.
- **Time on Site**: Users who convert generally spend more time on the website.
- **Social Shares**: The distribution appears similar, indicating that social shares may not be a strong differentiator.
- **Email Opens & Clicks**: Converted users generally have higher email interaction rates, suggesting that email engagement plays a role in conversion.  

These insights help us understand how different engagement behaviors impact conversions, guiding future marketing strategies.
